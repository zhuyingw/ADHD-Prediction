```python
import csv
import time
import numpy as np
import pandas as pd
import datetime as dt
from datetime import date
from datetime import datetime
from datetime import timedelta
import matplotlib.pyplot as plt
```

```python

def clean_duplicated():
    file1 = open("Activity_clean1.csv", "w", encoding="utf8")
    f1 = csv.writer(file1)
    file2 = open("Activity_clean2.csv", "w", encoding="utf8")
    f2 = csv.writer(file2)
    df=pd.read_csv("Activity1.csv")
    de=pd.read_csv("Activity2.csv")


    userid = df["Userid"].unique()

    for user in userid:
#         print(df[" senseStartTime"][df["Userid"]==user].unique(), len(df[" senseStartTime"][df["Userid"]==user].unique()), user)
        tmp=df.drop_duplicates(subset=[" senseStartTime", "Userid"], keep="first")
#         print(tmp)
        tmp.to_csv("Activity_clean1.csv", index=False)
    
    userid = de["Userid"].unique()
    
    for user in userid:
#         print(de[" senseStartTime"][de["Userid"]==user].unique(), len(de[" senseStartTime"][de["Userid"]==user].unique()), user)
        tmp=de.drop_duplicates(subset=[" senseStartTime", "Userid"], keep="first")
#         print(tmp)
        tmp.to_csv("Activity_clean2.csv", index=False)
    
    df1=pd.read_csv("Activity_clean1.csv")
    df2=pd.read_csv("Activity_clean2.csv")
    df_joined = pd.concat([df1,df2])
    df_joined.to_csv('Activity_joined.csv',index=False)
```

```python

clean_duplicated()

df = pd.read_csv("Activity_joined.csv")
#de = pd.read_csv("Qids_Score.csv")
df

dates_milli = df[" senseStartTime"].tolist()
date_mili_strip = [i.replace("'", "") for i in dates_milli]
date_final = [i.split("T")[0] for i in date_mili_strip]

date_final

converted_dates = []

for i in date_final:  # create a list fo dates from mili sec
    # s = int(i) / 1000.0
    # m = datetime.fromtimestamp(s).strftime('%Y-%m-%d')
    # # date=time.strftime('%Y-%m-%d', time.gmtime(int(i)))
    converted_dates.append(i)

df["Final Date"] = converted_dates  # new conlumn of dates only
df
# df.to_csv("Activity.csv", index=False)
```
